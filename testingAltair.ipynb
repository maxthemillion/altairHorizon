```python
states = alt.topo_feature(data.us_10m.url, feature='states')

# US states background
background = alt.Chart(states).mark_geoshape(
    fill='lightgrey',
    stroke='white'
).properties(
    width=500,
    height=300
).project('albersUsa')

# airport positions on background
points = alt.Chart(airports).transform_filter(
    (datum.country == 'USA') & (not (pd.isna(datum.city)))
).transform_aggregate(
    latitude='mean(latitude)',
    longitude='mean(longitude)',
    count='count()',
    groupby=['city', 'state', 'country']
).mark_circle().encode(
    longitude='longitude:Q',
    latitude='latitude:Q',
    size=alt.Size('count:Q', title='Number of Airports', scale=alt.Scale(domain=[1,10])),
    tooltip=['city:N','count:Q']
).properties(
    title='Number of airports in US'
)

alt.layer(background, points).configure_circle(
    radius=10,
    opacity=0.4
)
```

## Sales data

```python
sales = pd.read_csv('./sales.csv')
sales.date = pd.to_datetime(sales.date)
sales = sales[sales.date < '2015-11-01']
sales.head()
```

```python
sales['revenue'] = sales.item_cnt_day * sales.item_price
```

```python
item_ct = sales[['date', 'item_cnt_day', 'revenue']].groupby(sales.date).sum()
item_ct.tail()
```

```python
item_ct_week = item_ct.resample('W').sum()
```

```python
alt.Chart(item_ct_week.reset_index(), width=200).mark_bar().encode(
    x='month(date):O',
    y='sum(revenue)'
).facet(
    column='year(date)'
).configure_facet(
    spacing=2
).properties(
    title='Sum of revenue per week'
)
```

```python
item_ct_month = item_ct.resample('M').sum()
item_ct_month['change'] = (item_ct_month.revenue-item_ct_month.revenue.shift(12))/item_ct_month.revenue
```

```python
alt.Chart(item_ct_month.reset_index(), width=350).mark_bar().encode(
    alt.X('month(date):O', axis=alt.Axis(title='Time')),
    alt.Y('change:Q', axis=alt.Axis(title='Percentage change to previous year'))
).facet(
    column='year(date):O'
).properties(
    title='Change in revenue compared to previous year',
)
```

```python
items_per_store = sales.groupby([sales.shop_id, pd.Grouper(freq='M', key='date')]).sum()
```

```python
items_per_store = sales.groupby([sales.shop_id, sales.date]).sum()
```

```python
items_per_store = items_per_store.reindex( 
    pd.MultiIndex.from_product(
        [items_per_store.index.levels[0], 
        pd.date_range(np.min(items_per_store.index.levels[1]), np.max(items_per_store.index.levels[1]))], 
        names=['shop_id', 'date']), 
    fill_value=0)
```

```python
items_per_store = items_per_store.reset_index()
items_per_store = items_per_store.groupby([pd.Grouper(key='shop_id'), pd.Grouper(freq='M',key='date')]).sum()
```

```python
def calcChange(d):
    current = d.revenue
    previous = d.revenue.shift(12)
    return (current - previous)/previous
```

```python
def calcCtNorm(d):
    mean = np.mean(d.item_cnt_day)
    sd = np.std(d.item_cnt_day)
    return (d.item_cnt_day - mean)/sd
```

```python
def limit(d):
    v = d.item_cnt_day
    return (v - np.min(v))/(np.max(v)-np.min(v))
```

```python
items_per_store['change'] = 0
items_per_store['item_ct_norm'] = 0
items_per_store['limit'] = 0
for i in range(0,len(items_per_store.index.levels[0])):
    items_per_store.change.loc[i] = calcChange(items_per_store.loc[i]).values
    items_per_store.item_ct_norm.loc[i] = calcCtNorm(items_per_store.loc[i]).values
    items_per_store.limit.loc[i] = limit(items_per_store.loc[i]).values
```

```python
items_per_store.to_csv('./items.csv')
```
